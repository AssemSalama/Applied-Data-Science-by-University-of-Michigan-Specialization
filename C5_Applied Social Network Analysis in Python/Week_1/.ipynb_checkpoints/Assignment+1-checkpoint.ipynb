# Assignment 1 - Creating and Manipulating Graphs

Eight employees at a small company were asked to choose 3 movies that they would most enjoy watching for the upcoming company movie night. These choices are stored in the file `Employee_Movie_Choices.txt`.

A second file, `Employee_Relationships.txt`, has data on the relationships between different coworkers. 

The relationship score has value of `-100` (Enemies) to `+100` (Best Friends). A value of zero means the two employees haven't interacted or are indifferent.

Both files are tab delimited.

```python
import networkx as nx
import pandas as pd
import numpy as np
from networkx.algorithms import bipartite


# This is the set of employees
employees = set(['Pablo',
                 'Lee',
                 'Georgia',
                 'Vincent',
                 'Andy',
                 'Frida',
                 'Joan',
                 'Claude'])

# This is the set of movies
movies = set(['The Shawshank Redemption',
              'Forrest Gump',
              'The Matrix',
              'Anaconda',
              'The Social Network',
              'The Godfather',
              'Monty Python and the Holy Grail',
              'Snakes on a Plane',
              'Kung Fu Panda',
              'The Dark Knight',
              'Mean Girls'])


# you can use the following function to plot graphs
# make sure to comment it out before submitting to the autograder
def plot_graph(G, weight_name=None):
    '''
    G: a networkx G
    weight_name: name of the attribute for plotting edge weights (if G is weighted)
    '''
    %matplotlib notebook
    import matplotlib.pyplot as plt
    
    plt.figure()
    pos = nx.spring_layout(G)
    edges = G.edges()
    weights = None
    
    if weight_name:
        weights = [int(G[u][v][weight_name]) for u,v in edges]
        labels = nx.get_edge_attributes(G,weight_name)
        nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
        nx.draw_networkx(G, pos, edges=edges, width=weights);
    else:
        nx.draw_networkx(G, pos, edges=edges);
```

### Question 2

Using the graph from the previous question, add nodes attributes named `'type'` where movies have the value `'movie'` and employees have the value `'employee'` and return that graph.

*This function should return a networkx graph with node attributes `{'type': 'movie'}` or `{'type': 'employee'}`*

```python
B2 = answer_one()
B2.nodes()[:10]
```

```python
B2.edges()[:10]
```

```python
for node in B2.nodes():
    if node in employees:
        B2.add_node(node , type = 'employee')
    elif node in movies:
        B2.add_node(node , type = 'movie')
```

```python
print('num of edges =' ,len(B2.edges(data=True)))
print( 'num of nodes =' ,len(B2.nodes()))
plot_graph(B2)
```

```python
B2.nodes(data=True)
```

```python
def answer_two():
    B2 = answer_one()
    for node in B2.nodes() : 
        if node in employees:
            B2.add_node(node,type='employee')
        if node in movies :
            B2.add_node(node , type = 'movie')
    
    return B2
answer_two()
```

### Question 3

```python
B3 = answer_two()
we_pro_graph = bipartite.weighted_projected_graph(B3 , employees)
plot_graph(we_pro_graph)
```

```python
def answer_three():
    B3 = answer_two()
    we_pro_graph = bipartite.weighted_projected_graph(B3 , employees)

    return we_pro_graph
answer_three()
```

### Question 4

Suppose you'd like to find out if people that have a high relationship score also like the same types of movies.

Find the Pearson correlation ( using `DataFrame.corr()` ) between employee relationship scores and the number of movies they have in common. If two employees have no movies in common it should be treated as a 0, not a missing value, and should be included in the correlation calculation.

*This function should return a float.*

## 1: we'll use the relationship score between empolyees:

## 2: The number of movies in common we got from the previous Q: 


```python
G  = answer_three()
G.edges(data=True)
```

```python
G_df1 = pd.DataFrame(G.edges(data=True) ,
                    columns= ['From' , 'To' , 'common_movies'])
# removing the word 'weight'.
G_df1['common_movies'] = G_df1['common_movies'].map(lambda x: x['weight'])
G_df1.head(3)
```

### - we want to create a df in an interchangeable manner:
        ** where `Frida TO Pablo` =  `Pablo TO Frida `= 2 **

```python
G_df2 = pd.DataFrame(G.edges(data=True) ,
                    columns= ['To' , 'From' , 'common_movies'])
# removing the word 'weight'.
G_df2['common_movies'] = G_df2['common_movies'].map(lambda x: x['weight'])
G_df2.head(3)
```

```python
# finally , use concat() to put them on each other in one dataframe.
G_df = pd.concat([G_df1,G_df2] ,axis=0,ignore_index = True)

# check:
G_df.tail(11)
```

## 3: merge `relations` & `G_df` dataframes on `From` & `To` columns .

```python
final_df = pd.merge(relations , G_df , on=['From' ,'To'] , how='outer')
# convert any NaN to 0 in `common_movies` col.
final_df['common_movies'] = final_df['common_movies'].fillna(0)
final_df
```

```python
final_df['common_movies'].corr(final_df['rela_score'])
```

```python
def answer_four():
    relations = pd.read_csv('Employee_Relationships.txt' , sep='\t' ,
                header=None ,names= ['From' , 'To' , 'rela_score'])
    G  = answer_three()
    # 1------------------
    
    G_df1 = pd.DataFrame(G.edges(data=True) ,
                    columns= ['From' , 'To' , 'common_movies'])
    G_df1['common_movies'] = G_df1['common_movies'].map(lambda x: x['weight'])
    # 2------------------
    
    G_df2 = pd.DataFrame(G.edges(data=True) ,
                    columns= ['To' , 'From' , 'common_movies'])
    G_df2['common_movies'] = G_df2['common_movies'].map(lambda x: x['weight'])
    # 3------------------
    
    # finally , use concat() to put them on each other in one dataframe.
    G_df = pd.concat([G_df1,G_df2] ,axis=0,ignore_index = True)
    # merge `relations` & `G_df` .
    final_df = pd.merge(relations , G_df , on=['From' ,'To'] , how='outer')
    # convert any NaN to 0 in `common_movies` col.
    final_df['common_movies'] = final_df['common_movies'].fillna(0)
    # 4-------------------
    
    result = final_df['common_movies'].corr(final_df['rela_score'])
    return result

answer_four()
```
