### Question 6

Using the NetworkX function strongly_connected_component_subgraphs, find the subgraph of nodes in a largest strongly connected component. 
Call this graph G_sc.

*This function should return a networkx MultiDiGraph named G_sc.*

# Solution_1.
- Create a subgraph using the nodes in `largest_strong_conn_comp` using  the original directed graph `G` and the function `subgraph()`.

# Solution_2.

- using the `strongly_connected_component_subgraphs()` that creates a subgraph for each dic in the generated dics , then use `max(key=len)`.

Source:

https://networkx.org/documentation/networkx-1.10/reference/generated/networkx.algorithms.components.strongly_connected.strongly_connected_component_subgraphs.html

```python
G = answer_one()
# get a subgraph for each dic of nodes in the strongly_connected_component.
subgraphs = nx.strongly_connected_component_subgraphs(G)

max(subgraphs,key=len)
```

```python
def answer_six():
    G = answer_one()
    # get a subgraph for each dic of nodes in the strongly_connected_component.
    subgraphs = nx.strongly_connected_component_subgraphs(G)
    G_sc = max(subgraphs,key=len)

    return G_sc

answer_six()
```

### Question 11

Which node in G_sc is connected to the most other nodes by a shortest path of length equal to the diameter of G_sc?

How many nodes are connected to this node?


*This function should return a tuple (name of node, number of satisfied connected nodes).*

**- since we consider `shortest path of length =the diameter of G_sc` we'll use the periphery nodes to make sence which they are nodes that their largest distance between each of these and all other nodes = diameter.** 

```python
G_sc = answer_six()
d = nx.diameter(G_sc)
print(d)

peripheries = nx.periphery(G_sc)
print(peripheries)
```

```python
sp = nx.shortest_path_length(G_sc,'97')
sp
```

source : https://www.w3schools.com/python/ref_list_count.asp

The count() method returns the number of elements with the specified value.

- Here, we want to know how many nodes are in `distance = diameter=3` we'll do that using .Count() that count the number of recurring of element in a list.


```python
count = list(sp.values()).count(3)
count
```

```python
max_count = -1      # To avoid if there is no match count=0.
result_node = None  # empty To pack it. 

for node in peripheries:
    sp = nx.shortest_path_length(G_sc,node) # Do the operation.
    #count how many nodes are at the distance of 3 in terms of bfs_tree.
    count = list(sp.values()).count(d)
    if count > max_count:                     
        result_node = node
        max_count = count
        
(result_node , max_count)
```

```python
def answer_eleven():
    G_sc = answer_six()
    d = nx.diameter(G_sc)
    
    max_count = -1      # To avoid if there is no match count=0.
    result_node = None  # empty To pack it. 
    for node in peripheries:
        # Do the operation.
        sp = nx.shortest_path_length(G_sc,node)
        #count how many nodes are at the distance of 3 in terms of bfs_tree.
        count = list(sp.values()).count(d)
        if count > max_count:                     
            result_node = node
            max_count = count        
    
    return (result_node , max_count)
answer_eleven()
```

```python
nx.minimum_node_cut(G_sc,'38' ,'97')
```

**I have a question about the choice between node_connectivity and minimum_node_cut.**
- The node_connectivity works on an approximation algorithm and gives a tight lower bound, but not the actual/exact value for the number of nodes that need to be removed.


- Minimum_node_cut, on the other hand, gives an exact value by calculating all the paths between source and target.

```python
def answer_twelve():
    G_sc = answer_six()
    
    center= nx.center(G_sc)[0]
    node = answer_eleven()[0]
    
    nodes_to_remove = nx.minimum_node_cut(G_sc,'38' ,'97')
    their_num = len(nodes_to_remove)
    
    return their_num
answer_twelve()
```
